# univariant analysis 

```python
cat = ['STATUS','PRODUCTLINE','PRODUCTCODE','CUSTOMERNAME','PHONE','ADDRESSLINE1','CITY','POSTALCODE','COUNTRY','CONTACTLASTNAME','CONTACTFIRSTNAME','DEALSIZE']
num =[x for x in df.columns if x not in cat]
print(cat)
print(num)
```

```python
df['ORDERNUMBER'].value_counts()
```

# Univariate Analysis for categorical variable

```python
for catvar in cat:
    print(catvar,'  No of Levels:',df[catvar].nunique())
    print(df[catvar].value_counts())
    print('\n\n')
```

# Correlation Plot

```python
plt.figure(figsize=(10,10))
sns.heatmap(df[num].corr().round(2),
            annot=True, 
            mask = np.triu(df[num].corr(),1),
            cmap='plasma',
            vmin =-1, 
            vmax=1, 
            linewidths=2, 
            annot_kws={'size':10});
```
